```python
# basic example to explore cv2 libExample to 
import cv2 #opencv itself
import numpy as np # matrix manipulations

%matplotlib inline 
from matplotlib import pyplot as plt # for plotting
import pylab # for figure size 
pylab.rcParams['figure.figsize'] = (10.0, 10.0)
input_image=cv2.imread('Big_data_cartoon.jpg')
```

```python
plt.imshow(input_image)  # Show original diagram 
```

```python
flipped_code_0=cv2.flip(input_image,0) # show vertical flip
plt.imshow(flipped_code_0)
```

```python
flipped_code_1=cv2.flip(input_image,1) # show horizontal flip
plt.imshow(flipped_code_1)
```

```python
transposed=cv2.transpose(input_image)
plt.imshow(transposed)  #show tran
```
